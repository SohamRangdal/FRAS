```python
import cv2, os
import pandas as pd
import datetime
import time
import pymongo as py
import schedule
import time
import sys
from datetime import date






############################################for Morning Attendance########################################

def attendance_morg():
    connection = py.MongoClient("")


    database = connection['FRAS']
    collection = database['employee']
    collection1 = database['attendance_morg']
    collection2 = database['department']
    collection3 = database['role']
    collection5 = database['main_attendance']
    print("Database Connected")



    recognizer = cv2.face.LBPHFaceRecognizer_create()  # cv2.createLBPHFaceRecognizer()
    recognizer.read("Trainer/Trainer.yml")
    harcascadePath = "haarcascade_frontalface_default.xml"
    faceCascade = cv2.CascadeClassifier(harcascadePath)
    df = pd.read_csv('Employee_Details/Emp_Det.csv')
    cam = cv2.VideoCapture(0)
    font = cv2.FONT_HERSHEY_SIMPLEX
    col_names = ['Id', 'Name', 'Date', 'Time']
    attendance = pd.DataFrame(columns=col_names)
    while True:
        ret, im = cam.read()
        gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(gray, 1.2, 5)
        for (x, y, w, h) in faces:
            cv2.rectangle(im, (x, y), (x + w, y + h), (225, 0, 0), 2)
            Id, conf = recognizer.predict(gray[y:y + h, x:x + w])
            if (conf < 50):
                ts = time.time()
                date = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d')
                arialtamp = datetime.datetime.fromtimestamp(ts).strftime('%H:%M:%S')
                aa = df.loc[df['Id'] == Id]['Name'].values
                tt = str(Id) + "-" + aa
                attendance.loc[len(attendance)] = [Id, aa, date, arialtamp]
            else:
                Id = 'Unknown'
                tt = str(Id)
            if (conf > 75):
                noOfFile = len(os.listdir("Unknown_Images")) + 1
                cv2.imwrite("Unknown_Images/Image" + str(noOfFile) + ".jpg", im[y:y + h, x:x + w])
            cv2.putText(im, str(tt), (x, y + h), font, 1, (255, 255, 255), 2)
        attendance = attendance.drop_duplicates(subset=['Id'], keep='first')
        cv2.imshow('im', im)
        ts = time.time()
        date = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d')
        timeStamp = datetime.datetime.fromtimestamp(ts).strftime('%H:%M:%S')
        Hour, Minute, Second = timeStamp.split(":")
        fileName = "Attendance/Attendance_morg.csv"
        attendance.to_csv(fileName, index=False)
        
        localtime = time.localtime(time.time())
        if cv2.waitKey(1) == ord('q') or cv2.waitKey(1) == 27 or (localtime.tm_hour == 16 and localtime.tm_min == 0):
            time.sleep(10)
            break
    
    read_CSV = pd.read_csv("Attendance/Attendance_morg.csv")
    trans_data = read_CSV.to_dict('records')
    collection1.insert_many(trans_data)
    cam.release()
    cv2.destroyAllWindows()



    
########################################for Evening Attendance ########################################

def attendance_eve():
    connection = py.MongoClient("")


    database = connection['FRAS']
    collection = database['employee']
    collection1 = database['attendance_eve']
    collection2 = database['department']
    collection3 = database['role']
    print("Database Connected")



    recognizer = cv2.face.LBPHFaceRecognizer_create()  # cv2.createLBPHFaceRecognizer()
    recognizer.read("Trainer/Trainer.yml")
    harcascadePath = "haarcascade_frontalface_default.xml"
    faceCascade = cv2.CascadeClassifier(harcascadePath)
    df = pd.read_csv('Employee_Details/Emp_Det.csv')
    cam = cv2.VideoCapture(0)
    font = cv2.FONT_HERSHEY_SIMPLEX
    col_names = ['Id', 'Name', 'Date', 'Time']
    attendance = pd.DataFrame(columns=col_names)
    while True:
        ret, im = cam.read()
        gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(gray, 1.2, 5)
        for (x, y, w, h) in faces:
            cv2.rectangle(im, (x, y), (x + w, y + h), (225, 0, 0), 2)
            Id, conf = recognizer.predict(gray[y:y + h, x:x + w])
            if (conf < 50):
                ts = time.time()
                date = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d')
                arialtamp = datetime.datetime.fromtimestamp(ts).strftime('%H:%M:%S')
                aa = df.loc[df['Id'] == Id]['Name'].values
                tt = str(Id) + "-" + aa
                attendance.loc[len(attendance)] = [Id, aa, date, arialtamp]
            else:
                Id = 'Unknown'
                tt = str(Id)
            if (conf > 75):
                noOfFile = len(os.listdir("Unknown_Images")) + 1
                cv2.imwrite("Unknown_Images/Image" + str(noOfFile) + ".jpg", im[y:y + h, x:x + w])
            cv2.putText(im, str(tt), (x, y + h), font, 1, (255, 255, 255), 2)
        attendance = attendance.drop_duplicates(subset=['Id'], keep='first')
        cv2.imshow('im', im)
        ts = time.time()
        date = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d')
        timeStamp = datetime.datetime.fromtimestamp(ts).strftime('%H:%M:%S')
        Hour, Minute, Second = timeStamp.split(":")
        fileName = "Attendance/Attendance_eve.csv"
        attendance.to_csv(fileName, index=False)
        
        
        localtime = time.localtime(time.time())
        if cv2.waitKey(1) == ord('q') or cv2.waitKey(1) == 27 or (localtime.tm_hour == 18 and localtime.tm_min == 15):
            time.sleep(10)
            break
    
    read_CSV = pd.read_csv("Attendance/Attendance_eve.csv")
        
    
    trans_data = read_CSV.to_dict('records')
    collection1.insert_many(trans_data)
    cam.release()
    cv2.destroyAllWindows()
   



"""
##############################################main Attendance############################################



def main_attend():
    connection = py.MongoClient("")
    
    database = connection['FRAS']
    collection1 = database['attendance_morg']
    collection2 = database['attendance_eve']
    collection = database['main_attendance']
    print("database connected")
   
    today = date.today()
    result_eve = collection2.find({"Date": today.strftime('%Y-%m-%d')})
    result_morg = collection1.find({"Date": today.strftime('%Y-%m-%d')})
    
    for results in result_morg:
        for results2 in result_eve:
            temp = results["Id"]
            temp2 = results2["Id"]
            temp3 = results["Name"]
            if temp == temp2 :
                collection.insert({"Id": temp2, "Name": temp3, "Date": today.strftime('%Y-%m-%d')})
                print("inserted")
                
            else:
                print("not inserted")
    
    
   
    
main_attend()
"""
schedule.every().day.at("").do(attendance_morg)        #give the timings in .at("set time")
schedule.every().day.at("").do(attendance_eve)         #give the timings in .at("set time")

#schedule.every().day.at("").do()
while True:
    schedule.run_pending()
    time.sleep(1)


```
